# Phase 3: Data Cleaning

## Objective

The goal of this phase is to prepare the raw customer churn dataset for Exploratory Data Analysis (EDA) and Machine Learning.

Real-world datasets often contain:
- Missing values
- Duplicate records
- Irrelevant columns
- Data leakage
- Incorrect data types
- Inconsistent values

Cleaning the data improves data quality and helps build more reliable machine learning models.

In this notebook, we will:

- Create a copy of the original dataset
- Remove unnecessary columns
- Identify and remove data leakage
- Handle missing values
- Verify data consistency
- Save the cleaned dataset

```python
import pandas as pd
import numpy as np
```

# Step 2: Load the Dataset

Load the original customer churn dataset from the data folder.

```python
df = pd.read_csv("../data/raw/telco.csv")
```

# Step 3: Create a Copy of the Dataset

Instead of modifying the original dataset directly, we create a copy.

This preserves the raw dataset in case we need to refer back to it later.

```python
df_clean = df.copy()
```

# Step 4: Check Dataset Shape

Understanding the number of rows and columns before cleaning helps us track how the dataset changes during preprocessing.

```python
print(df_clean.shape)
```

# Step 5: Remove Data Leakage Columns

Some columns contain information that would only be available after the customer has already churned.

Using these columns during model training would cause data leakage and lead to unrealistically high model performance.

Columns removed:

- Churn Score
- Churn Category
- Churn Reason

```python
leakage_columns = [
    "Churn Score",
    "Churn Category",
    "Churn Reason"
]

df_clean.drop(columns=leakage_columns, inplace=True)
```

# Step 6: Remove Identifier Column

Customer ID uniquely identifies each customer but does not help predict churn.

Such columns should not be used as machine learning features.

```python
df_clean.drop(columns=["Customer ID"], inplace=True)
```

# Step 7: Remove Constant Columns

Columns that contain only one unique value provide no useful information for prediction.

Removing them reduces unnecessary features.

Columns removed:

- Country
- State
- Quarter

```python
constant_columns = [
    "Country",
    "State",
    "Quarter"
]

df_clean.drop(columns=constant_columns, inplace=True)
```

# Step 8: Verify Dataset Shape

Check how many columns remain after removing unnecessary features.

```python
print(df_clean.shape)
```

# Step 9: Verify Removed Columns

Ensure that all unwanted columns have been successfully removed.

```python
df_clean.columns
```

# Step 10: Check Missing Values

Identify columns that still contain missing values.

```python
df_clean.isnull().sum().sort_values(ascending=False)
```

# Step 11: Investigate Missing Internet Type Values

Internet Type is expected to be missing for customers who do not have internet service.

We verify this relationship before handling missing values.

```python
pd.crosstab(
    df_clean["Internet Service"],
    df_clean["Internet Type"],
    dropna=False
)
```

# Step 12: Investigate Missing Offer Values

The Offer column contains missing values.

Before replacing them, we examine the distribution to understand whether missing values indicate that no promotional offer was provided.

```python
df_clean["Offer"].value_counts(dropna=False)
```

# Step 13: Check Duplicate Records

Duplicate rows can introduce bias into the model.

Verify that the dataset does not contain duplicate customer records.

```python
df_clean.duplicated().sum()
```

# Step 14: Check Data Types

Verify the data types of all remaining columns before proceeding with preprocessing.

```python
df_clean.info()
```

# Summary

In this notebook, we:

- Created a working copy of the dataset.
- Removed columns that could cause data leakage.
- Removed identifier and constant-value columns.
- Investigated missing values.
- Verified duplicate records.
- Checked the dataset structure and data types.

The cleaned dataset is now ready for further preprocessing and Exploratory Data Analysis (EDA).

# Step 15: Save the Cleaned Dataset

Save the cleaned dataset so it can be reused in future notebooks without repeating the cleaning process.

```python
df_clean.to_csv("../data/processed/telco_clean.csv", index=False)
```
